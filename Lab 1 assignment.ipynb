## Task 1

Use `SymPy` to solve the differential equation $y' = (2y-3y^2)\sin(x)$, with $y(0)=0.5$, and plot the solution.

```python
import sympy as sym
from IPython.display import display_latex
import sympy.plotting as sym_plot

# Solving the equaton
x = sym.symbols('x')
y = sym.Function('y')
eq = sym.Eq(y(x).diff(x), (2*y(x)-3*y(x)*y(x))*sym.sin(x))
display_latex(eq)
sol = sym.dsolve(eq, y(x), ics={y(0):0.5})
display_latex(sym.simplify(sol))

# Plotting the graph
sym_plot.plot(sol.rhs, xlabel = 'x', ylabel = 'y')
```

## Task 2

Use `SciPy`'s `odeint` function to solve the system of equations

\begin{align*}
\frac{dx}{dt}&=xy -2.1x \\ \frac{dy}{dt} &= 0.8y - 0.4xy
\end{align*}

Produce a plot of the solutions for $0\leq t\leq 20$ with initial conditions $x(0)=0.5$ and $y(0) \in \{1, 3, 5, \ldots, 19\}$.

How many curves do you expect to see plotted?

Qualitatively, what happens to the solution, $(x(t),y(t))$, as $y(0)$ becomes large?

```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Define the equation
def dy_dx(Y, t):
    x, y = Y
    return [x*y - 2.1*x, 0.8*y - 0.4*x*y]

t = np.linspace(0, 20, 1000)
x0 = 0.5

# Record number of plot
n = 0

for y0 in np.arange(1, 20, 2):
    X0 = [x0, y0]
    Xsol = odeint(dy_dx, X0, t)
    x, y = Xsol[:, 0], Xsol[:, 1]
    plt.plot(Xsol[:, 0], Xsol[:, 1], label=f'y(0)={y0}')
    n += 1

# Plot the graph
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.legend()
plt.grid()
plt.show()

print(f'Number of curves is {n}')

```

The curves tend to move in a certain path to form a loop.
